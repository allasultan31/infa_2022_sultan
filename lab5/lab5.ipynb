Построить графики для Chromo.txt, mutant.txt и wild_type.txt из лабораторной №2. Для Chromo.txt построить только отрезок от 4 до 7 минут.

```python
import pandas as pd

#chromo
df = pd.read_csv('Chromo.txt', sep = '\t')
df1 = df[(df['Time (min)']>=4) & (df['Time (min)']<=7)]
df1.plot('Time (min)','Value (mAU)', ylabel = 'Value (mAU)', title = 'Chromo: 4-7 min')

#mutant
df2 = pd.read_csv('mutant.txt', sep = '\t')
df2.plot('Concentration, nM', 'Response', ylabel = 'Response', title = 'mutant')

#wild_type
df3 = pd.read_csv('wild_type.txt', sep = '\t')
df3.plot('Concentration, nM', 'Response', ylabel = 'Response', title = 'wilde_type')
```
