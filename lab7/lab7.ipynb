## Упражнение 1

Для каждого уровня образования родителей (parental level of education) постройте как subplot'ы:
- точечный график результатов экзамена по чтения и экзамена по математике
- точечный график результатов экзамена по чтения и экзамена по письму

```python
import matplotlib.pyplot as plt
import pandas as pd

students = pd.read_csv("StudentsPerformance.csv")

s_h = students[students["parental level of education"]== "some high school"]
b_d = students[students["parental level of education"]== "bachelor's degree"] 
s_c = students[students["parental level of education"]== "some college"] 
m_d = students[students["parental level of education"]== "master's degree"] 
a_d = students[students["parental level of education"]== "associate's degree"] 
h_s = students[students["parental level of education"]== "high school"] 
 
for df in (s_h, b_d, s_c, m_d, a_d, h_s):
    plt.scatter(df["reading score"], df["math score"])
level_types = ["some high school", "bachelor's degree" "some college", "master's degree", "associate's degree", "high school"]
plt.legend(level_types)
plt.title('parental level of education,\n maths&reading scores')
plt.xlabel('reading score')
plt.ylabel('math score')
plt.show()

for df in (s_h, b_d, s_c, m_d, a_d, h_s):
    plt.scatter(df["reading score"], df["writing score"])
level_types = ["some high school", "bachelor's degree" "some college", "master's degree", "associate's degree", "high school"]
plt.legend(level_types)
plt.title('parental level of education,\n writing&reading scores')
plt.xlabel('reading score')
plt.ylabel('writing score')
plt.show()
```

## Упражнение 2

Для каждого уровня образования родителей и для каждого экзамена (чтение, письмо, математика) постройте ящики с усами.

```python
import matplotlib.pyplot as plt
import pandas as pd

students = pd.read_csv("StudentsPerformance.csv")


s_h = students[students["parental level of education"]== "some high school"]
b_d = students[students["parental level of education"]== "bachelor's degree"] 
s_c = students[students["parental level of education"]== "some college"] 
m_d = students[students["parental level of education"]== "master's degree"] 
a_d = students[students["parental level of education"]== "associate's degree"] 
h_s = students[students["parental level of education"]== "high school"] 


exams = ["math score", "reading score", "writing score"]
for i in range(3):
    plt.boxplot((s_h[exams[i]], b_d[exams[i]], s_c[exams[i]], m_d[exams[i]], a_d[exams[i]], h_s[exams[i]]), 
                labels=level_types)
    plt.title('parental level of education \n' + exams[i])
    plt.xlabel('parental level of education')
    plt.ylabel(exams[i])
    plt.show()
```

Вам дан датасет miceprotein.csv

<hr>Уровни экспрессии 77 белков измерены в коре головного мозга 8 классов контрольных мышей и мышей с синдромом Дауна, подвергнутых контекстному обусловливанию страха - задаче, используемой для оценки ассоциативного обучения.

Набор данных состоит из уровней экспрессии 77 белков/белковых модификаций, которые давали детектируемые сигналы в ядерной фракции коры головного мозга. Имеется 38 контрольных мышей и 34 трисомные мыши (синдром Дауна), всего 72 мыши. В экспериментах было зарегистрировано 15 измерений каждого белка на образец/мышь. Таким образом, для контрольных мышей имеется 38x15, или 570 измерений, а для трисомных мышей - 34x15, или 510 измерений. Всего набор данных содержит 1080 измерений для каждого белка. Каждое измерение можно рассматривать как независимый образец/мышь.

Восемь классов мышей описаны на основе таких характеристик, как генотип, поведение и лечение. Согласно генотипу, мыши могут быть контрольными или трисомными. Согласно поведению, одни мыши были стимулированы к обучению (контекст-шок), а другие нет (шок-контекст), и для того, чтобы оценить эффект препарата мемантин в восстановлении способности к обучению у трисомных мышей, некоторым мышам вводили препарат, а другим нет.

Для датасета miceprotein.csv постройте ящики с усами для уровней экспрессии белков `DYRK1A_N`, `SOD1_N` для каждого класса образцов. ⚠Некоторые данные отсутствуют. Для выбрасывания NaN-значений используйте метод `dropna`.

```python
import pandas as pd;
import matplotlib.pyplot as plt;

miceprotein = pd.read_csv("miceprotein.csv", na_values = "?").dropna()  #чтение файла с заменой ? на Nan (пустое значение)
                                                                        #и удаление пустых значений из набора данных 

class_A = miceprotein[miceprotein["class"] =="'c-CS-m'"]     #выборка строк соответствующего класса
class_B = miceprotein[miceprotein["class"] =="'c-SC-m'"]
class_C = miceprotein[miceprotein["class"] =="'c-CS-s'"]
class_D = miceprotein[miceprotein["class"] =="'c-SC-s'"]
class_E = miceprotein[miceprotein["class"] =="'t-CS-m'"]
class_F = miceprotein[miceprotein["class"] =="'t-SC-m'"]
class_G = miceprotein[miceprotein["class"] =="'t-CS-s'"]
class_H = miceprotein[miceprotein["class"] =="'t-SC-s'"]
title_class = ['c-CS-m',"c-SC-m" ,"c-CS-s",'c-SC-s','t-CS-m','t-CS-m','t-CS-m','t-SC-s'] 

#первый график
plt.figure('экспрессия белков DYRK1A_N')    
plt.boxplot((class_A["DYRK1A_N"].astype('float'), #приведение значений к числовому виду astype('float')
            class_B["DYRK1A_N"].astype('float'),
            class_C["DYRK1A_N"].astype('float'),
            class_D["DYRK1A_N"].astype('float'),
            class_E["DYRK1A_N"].astype('float'),
            class_F["DYRK1A_N"].astype('float'),
            class_G["DYRK1A_N"].astype('float'),
            class_H["DYRK1A_N"].astype('float')), labels=title_class)

plt.title('экспрессия белков DYRK1A_N')
plt.xlabel('классы мышей')
plt.ylabel('набор данных')
plt.grid(True)
    

#второй график
plt.figure('экспрессия белков ITSN1_N')    
plt.boxplot((class_A["ITSN1_N"].astype('float'),
            class_B["ITSN1_N"].astype('float'),
            class_C["ITSN1_N"].astype('float'),
            class_D["ITSN1_N"].astype('float'),
            class_E["ITSN1_N"].astype('float'),
            class_F["ITSN1_N"].astype('float'),
            class_G["ITSN1_N"].astype('float'),
            class_H["ITSN1_N"].astype('float')), labels=title_class)

plt.title('экспрессия белков ITSN1_N')
plt.xlabel('классы мышей')
plt.ylabel('набор данных')
plt.grid(True)

plt.show()
```
