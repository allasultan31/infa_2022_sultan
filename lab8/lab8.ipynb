## Упражнение 3

Для каждого файла из упражнения №1 постройте точечный график (первый столбец - абсциссы, остальные - ординаты различных серий) и сохраните как `png`. Все графики должны оказаться в специально созданной скриптом папке.

```python
import os
import pandas as pd;
import matplotlib.pyplot as plt;

colors = ['purple', 'red', 'green', 'black', 'yellow', 'cyan'] #список цветов для отображения точек на графике

root_path = 'C:\\Users\\User\\Desktop\\lab8'           
source_folder = 'Source'          #папка с исходными файлами
imag_folder = 'Images'            #папка с рисунками

source_path = os.path.join(root_path, source_folder)        
imag_path = os.path.join(root_path, imag_folder)            

if not(os.path.exists(imag_path)):                          #если нет папки для рисунков, то создаем её   
    os.mkdir(imag_path)                                     

#функция чтения файла
def readFile(index):
    currentName = "{:02d}".format(index);           #формирование имени файла, в зависимости от индекса
    csv_name = currentName + '.csv.txt'             
    csv_file = os.path.join(source_path, csv_name)  
    pngName = currentName + '.png'                     
    print(csv_file)                                
    df = pd.read_csv(csv_file, header=None)         
    paintFile(df, pngName)                          

#функция формирования графика
def paintFile(df, name):
    columns_count = df.shape[1]   #определение количества столбцов в DataFrame
    
    list_X = df[0]     #первый столбец - абсцисса

    plt.title(name)
    plt.xlabel('X')
    plt.ylabel('Y')
    
    j=1;
    while j < columns_count:                             
        plt.scatter(list_X, df[j], color = colors[j-1])  #df[j] - ординаты различных серий
        j += 1                                           
    
    png_file = os.path.join(imag_path, name)             
    plt.savefig(png_file)                                 
    plt.show()                                          


#графики
i = 0
while i < 42:
    df = readFile(i)   
    i += 1
```
