## Упражнение 2

В файле `chromo.txt` находится хроматограмма - сигнал детектора в зависимости от времени. Продемонстрируйте данные на графике. Столбцы разделены табуляцией - `\t`. Разрешается (и приветствуется) просмотр исходных файлов.

```python
import numpy as np
import matplotlib.pyplot as plt
plt.figure(figsize=(10,10))
Chromo = open ("/Users/User/Downloads/Spring/Lab2/Chromo.txt")

value_x = []
value_y = []
for line in Chromo: 
    if line.split("\t")[0] != "Time (min)":
        value_x.append(float(line.split("\t")[0]))
        value_y.append(float(line.split("\t")[1]))

plt.plot(value_x, value_y) 
plt.title('Хроматограмма')
plt.xlabel('время, мин')
plt.ylabel('значение сигнала, mAU')

plt.show()
```

## Упражнение 3

Напишите функцию, которая определяет коэффициент корреляции. Посчитайте с её помощью коэффициент корреляции для данных, показанных в примере для точечного графика (3 набора), а также из упражнения 1 (2 набора)

```python
import numpy as np
mutant = open (r'C:\Users\User\Downloads\Spring\Lab2\mutant.txt')

def reading_x (name):
    f=open(name, 'r')
    Flag=True
    x=[]
    for line in f:
        if Flag==False:
            a=line.split()
            x.append(float(a[0]))
        Flag=False
    f.close()
    return x
def reading_y (name):
    f=open(name, 'r')
    Flag=True
    y=[]
    for line in f:
        if Flag==False:
            a=line.split()
            y.append(float(a[1]))
        Flag=False
    f.close()
    return y
x=reading_x('mutant')
y=reading_y('mutant')
n=len(x)

def r (x,y):
    Dx= np.std (x)
    Dy= np.std (y)
    for i in range (n):
        a=((x[i]-np.mean(x))/Dx)*((y[i]-np.mean(y))/Dy)
        S=np.sum(a)
        r=1/n*S
    return r

r= r(x,y)

print (r)
```
